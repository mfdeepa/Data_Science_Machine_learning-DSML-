```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn

```

```python
actions = {}
```

* Created an empty set actions to note the activities

```python
data_dirs = ["./EMG Physical Action Data Set/sub1/Aggressive/txt",
             "./EMG Physical Action Data Set/sub1/Normal/txt"]
```

* We can pre-process to all the subjects, for now we will limit to subject 1, and load only the subject 1's data.

```python
ind = 0
data = pd.DataFrame()

for dirs in data_dirs :

  for files in os.listdir(dirs):

    with open(os.path.join(dirs, files), "r") as f:

      temp = pd.read_csv(f.name,
                        sep = "\t",
                        header = None,
                        names = ["ch" + str(i) for i in range(1, 9)] # 8 input channels
                        )

      # chunking using Max of every 10 sequential values.
      temp_chunked = pd.DataFrame()

      for i in range(0, len(temp), 10):
        temp_chunked = temp_chunked.append(temp.iloc[i:i+10].max(), ignore_index = True)

      labels = [files[:-4] for i in range(len(temp_chunked))] # remove the last 4 characters=".txt" from the filename
      actions[files[:-4]] = ind

      temp_chunked["Action"] = labels

      data = pd.concat([data, temp_chunked])

      ind+=1

print(actions)
```

* So, here in the above code, in for loops we have opened every file in the folder mentioned in the list "data_dirs"
* Then we are opening every file with read only permission
* Then we are creating a csv file with tab as seperator, and naiming the channels as ch1,ch2...

* Now, to reduce the data set what we can do is, instead of considering all 10 data points in 1 milli second, we will consider only the maximum value in that milli second.
 * This is called  chunking or max operation.


 * Now we applied this max operation on the temp data which we created and considered only the maximum observation in 10 observations in 1 milli second and appended in temp_chunked

### K-fold cv

* This method is useful especially when we have small data set sizes, as if you split the whole data set in the ratio 6:2:2 ratio respectively for train, test , cross validation the model may not train properly
* So, using k-fold cv we can split the data set into train, test dataset and get the cross validation set out of train data instead of splitting





<img src='https://drive.google.com/uc?id=1tOKYuZQvjcj_u9zW1UHu1STOUUQcbqdq' >

**How K-fold cv works?**

For example, let's assume, we want to find the depth of the tree. So, let initial value of depth be 4. (contd)
Let the k value be 5 for our case
* First we devide the whole data set to k (5) parts
* For our depth 4, we do k (5) iterations on the data set
* In out 1st iteation we use 4 parts as train data , 1 part of the data as cross validation data and calculate the cross validation score for that data. Let that be CV1 for the same depth
* In the 2nd iteration we use another chunk of data as cross validation data and calculate CV2 .
* Repeat this in every iteration.
* So at the end we will have 5 cross validation scores, take the average of all to get the cross validation score for depth=4.

**What is the purpose of K-fold cv?**
* It's main purpose is to find a hyper-parameter
* As every point will be used in either cross validation or train data set, the score obtained will be more robust

* We will do this for various values of hyper-parameters to choose the optimal one basing on the average cross validation data
* Here the **disadvantage** is as we are training more models the speed decreases

* These curves are called **learning curves**
* These curves give us the idea how the trtain score or validation score changes with respect to the size of the data
* When the size of the data is less, both the training and validation curves diverge and they converge as the size of the dataset increase.
Let's see the code for the same.

```python
# Learning Curves
from sklearn.model_selection import learning_curve

def plot_learning_curve(estimator, X, Y, title):

  train_sizes, train_scores, test_scores, _, _ = learning_curve(estimator,
                                                                        X,
                                                                        Y,
                                                                        return_times=True
                                                                        )

  fig, axes = plt.subplots(1, 1, figsize = (10, 5))

  axes.set_title(title)
  axes.plot
  axes.set_xlabel("Training examples")
  axes.set_ylabel("Score")


  train_scores_mean = np.mean(train_scores, axis=1)
  train_scores_std = np.std(train_scores, axis=1)
  test_scores_mean = np.mean(test_scores, axis=1)
  test_scores_std = np.std(test_scores, axis=1)

  # Plot learning curve
  axes.grid()
  axes.fill_between(
      train_sizes,
      train_scores_mean - train_scores_std,
      train_scores_mean + train_scores_std,
      alpha=0.1,
      color="r",
  )
  axes.fill_between(
      train_sizes,
      test_scores_mean - test_scores_std,
      test_scores_mean + test_scores_std,
      alpha=0.1,
      color="g",
  )
  axes.plot(
      train_sizes, train_scores_mean, "o-", color="r", label="Training score"
  )
  axes.plot(
      train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
  )
  axes.legend(loc="best")

  plt.show()
```

We create a function plot_learning_curves and give it the model estimator, which is our best model(decision tree), and also the values of x, y and label it using the learning curve function.

**What is the highlighted area in the graph?**

* When we do k-fold cross validation we have k cross validation errors and k train errors, the line represents the mean accuracy of the models and the width(highlighted area) is the function of standard deviation
* It is basically the difference between mean plus standard deviation and mean minus standard deviation.
* It gives us the understanding of the difference in the errors of each model in the k-fold cv that therefore tells us about the varaince of the data

## Cascading

Lets, assumme we are to detect a fraudulent transaction or not

 Let the dataset be $D_1$ which will be imbalanced, and $y=1$ for fraudulent transaction

* For a query point $x_q$, we will pass this point through the first model $M_1$
* If the probability of $y$ being 1 is extremely low, say $<0.001$ then we consider that as not fraudulent, let this data be $D_1'$.
* The rest of the points ($D_1-D_1'$) which we are not sure about will be passed through the next model $M_2 $ which will be more stricter in finding fraud like <0.01.
* In the first model we are just removing all the genuine customers, in second model we are trying to find the may be fraudalent points from 2nd data set, we contimue doing this **cascading**
* Every model is trained oon different datasets
* If even after all these models, we are not sure there will be a human at last to verify the same

## Stacking

Lets assume we have a data set of $n$ poits.
* We train $m$ **individual** models on this data set, these models can be different types of models likemone can be Decision tree, one can GBDT other can be Random forest etc.
  * Let these base learners be $c_1,c_2,....c_n$
* Now we will take the outputs of each models as predictions ($p_1,p_2,...p_n$) of each model and train a Meta-Classifier on these outputs.
 * The Meta-classifier can be any model
* And this Meta-classifier gives the final output of the data

* Here what we are doing is that we are taking the outputs of the perfectly build models and stacking them together to train a Meta-classifier to get the final output

This is implemented using the **StackingClassifier** library from **miextend.classifier** module
* In the code we import all the libraries first
* Then **create the base classfiers** and pass these as **inputs to stacking classfier** and use **set any model as Meta-Classifier.**
* Now, **we train the model** and it's done
